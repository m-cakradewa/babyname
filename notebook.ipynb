```python
import pandas as pd
import numpy as np

# Define the years
years = list(range(1970, 2026, 5))  # 1970, 1975, ..., 2025

# Function to generate popularity numbers based on category
def generate_popularity(name, category):
    popularity = []
    for y in years:
        if category == 'classic':
            # Classic: steady, slightly decreasing
            base = np.random.randint(40, 70)
            pop = max(0, base + np.random.randint(-10, 10))
        elif category == 'modern':
            # Modern: very low early years, increasing after 2000
            if y < 2000:
                pop = np.random.randint(0, 5)
            else:
                pop = np.random.randint(20, 80)
        elif category == 'uprising':
            # Uprising: moderate early, dip, then rise again
            if y < 1995:
                pop = np.random.randint(15, 35)
            elif y < 2010:
                pop = np.random.randint(5, 20)
            else:
                pop = np.random.randint(30, 70)
        elif category == 'rare':
            # Rare: mostly low popularity
            pop = np.random.randint(0, 15)
        elif category == 'steady':
            # Steady: moderate, fairly consistent
            pop = np.random.randint(20, 50)
        popularity.append(pop)
    return popularity

# Prepare dataframe
data = {'Name': [], 'Year': [], 'Popularity': []}

# Combine all lists with their category
all_names = [
    (classic_names, 'classic'),
    (modern_names, 'modern'),
    (uprising_names, 'uprising'),
    (rare_names, 'rare'),
    (steady_names, 'steady')
]

for name_list, category in all_names:
    for name in name_list:
        pops = generate_popularity(name, category)
        for year, pop in zip(years, pops):
            data['Name'].append(name)
            data['Year'].append(year)
            data['Popularity'].append(pop)

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('name_popularity.csv', index=False)

print("CSV file 'name_popularity.csv' generated successfully!")
```
